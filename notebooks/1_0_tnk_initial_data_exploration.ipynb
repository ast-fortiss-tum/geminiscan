# Imports

```python
from sklearn.metrics import precision_recall_fscore_support
import numpy as np
import pandas as pd
```

# Calc base F1 Score

```python
states = pd.read_csv('data/SS.csv')
```

```python
states = states[states['HUMAN_CLASSIFICATION'] != 0]
```

```python
states["pred"] = 2
```

```python
states = states.loc[states['appname'] == "petclinic"]
states
```

```python
precision, recall, fbeta_score, support = precision_recall_fscore_support(states["HUMAN_CLASSIFICATION"], states["pred"], zero_division=1)
print(f"precision: {precision} recall: {recall} f1: {fbeta_score}")
```

```python
precision, recall, fbeta_score, support = precision_recall_fscore_support(states["HUMAN_CLASSIFICATION"], states["pred"], zero_division=1, average="macro")
print(f"precision: {precision} recall: {recall} f1: {fbeta_score}")
```
